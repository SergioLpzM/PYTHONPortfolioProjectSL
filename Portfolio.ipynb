```python
 # PREP: Import Libraries and Load Dataset

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

mpg_data = sns.load_dataset('mpg')

mpg_data.head()
```

```python
#QUESTION 1: Create a Visual Showing the Relationship Between Horsepower and Fuel Economy

plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower', y='mpg', data=mpg_data, color='darkblue')


plt.title('Horsepower vs Fuel Economy (mpg)', fontsize=16)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Fuel Economy (mpg)', fontsize=12)
plt.show()

# OBSERVATION: As horsepower increases, fuel economy (mpg) decreases. This suggests that more powerful cars are generally less fuel-efficient.
```

```python
#QUESTION 2: Create a table summarizing fuel economy by country of origin

# Summarizing fuel economy by origin
fuel_economy_by_origin = mpg_data.groupby('origin')['mpg'].mean().reset_index()

fuel_economy_by_origin
```

```python
# Creating a bar plot to visualize the summarised data

plt.figure(figsize=(8, 6))
sns.barplot(x='origin', y='mpg', data=fuel_economy_by_origin, palette='Set3')

plt.title('Average Fuel Economy by Country of Origin', fontsize=16)
plt.xlabel('Country of Origin', fontsize=12)
plt.ylabel('Average Fuel Economy (mpg)', fontsize=12)
plt.show()

# OBSERVATION: This shows that Japan produces cars with better fuel economy on average.
```

```python
# QUESTION 3: Create a Visual Showing How Fuel Economy Has Changed Over Time

plt.figure(figsize=(8, 6))
sns.lineplot(x='model_year', y='mpg', data=mpg_data, ci=None, color='green')

plt.title('Fuel Economy Over Time (Model Year)', fontsize=16)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Fuel Economy (mpg)', fontsize=12)
plt.show()

# OBSERVATION: We may observe that fuel economy has improved over the years, especially in response to energy crises or changing regulations.
```

```python
# QUESTION 4: Create Two Other Visuals or Tables Related to Fuel Economy

plt.figure(figsize=(8, 6))
sns.boxplot(x='origin', y='mpg', data=mpg_data, palette='Set2')

plt.title('Fuel Economy Distribution by Country of Origin', fontsize=16)
plt.xlabel('Country of Origin', fontsize=12)
plt.ylabel('Fuel Economy (mpg)', fontsize=12)
plt.show()

# OBSERVATION: The boxplot shows the spread of mpg values for each country, with Japan at the top.
```

```python
# QUESTION 5: Create Two Other Visuals or Tables Related to Fuel Economy

plt.figure(figsize=(8, 6))
sns.violinplot(x='cylinders', y='mpg', data=mpg_data, palette='Set2')

plt.title('Fuel Economy by Number of Cylinders', fontsize=16)
plt.xlabel('Cylinders', fontsize=12)
plt.ylabel('Fuel Economy (mpg)', fontsize=12)
plt.show()

# OBSERVATION: Cars with fewer cylinders tend to have better fuel economy, which is evident from the distribution of mpg.
```
